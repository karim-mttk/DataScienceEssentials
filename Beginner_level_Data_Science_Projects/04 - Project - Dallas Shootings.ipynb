## Step 3: Analyze
- Feature selection
- Model selection
- Analyze data

````python
subject_incidents.columns.tolist()
````

````python

dataset = subject_incidents[['race', 'subject_statuses', 'latitude', 'longitude']].dropna()
````

````python
len(dataset)
````

### Step 3.b: How to visualize features
- visualizing the two features: `race` and `subject_statuses`

### Step 3.b.1: Converting column
```

````python
print(dataset['race'].unique())
````

````python
mapping = {'B': 'blue', 'W': 'yellow', 'A': 'red', 'L': 'cyan'}
````

````python
dataset['race'] = dataset['race'].apply(lambda x: mapping[x])
dataset.head()
````

### Step 3.b.2: Converting column

````python
mapping = {'Deceased': 1000, 'Injured': 500, 'Shoot and Miss': 250}
dataset['subject_statuses'] = dataset['subject_statuses'].apply(lambda x: mapping.get(x, 100))
````

### Step 3.c: Visualizing the data

````python
dataset.plot.scatter(x='longitude', y='latitude', s='subject_statuses', c='race', alpha=.2)
plt.title('Figure 2: Subject statuses based on race scatter plot')
````

### Step 4.a Improved visualization
- Visualizing it on a map
    -  [folium](https://python-visualization.github.io/folium/)
    -  [quick guide](https://python-visualization.github.io/folium/quickstart.html)

````python
#!pip install folium
import folium
````

````python
dallas = folium.Map(location=[32.7767, -96.7970], zoom_start=10)
````

````python
dallas
````

````python
for _, row in dataset.iterrows():
    folium.CircleMarker(
    location=[row['latitude'], row['longitude']],
    radius=row['subject_statuses'] / 100,
    color=row['race'],
    fill=True,
    fill_color=row['race'],
).add_to(dallas)
````
