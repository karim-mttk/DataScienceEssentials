### Grouping Data with Pandas

```python
import pandas as pd
import numpy as np

df = pd.read_csv('datasets/census.csv')
df = df[df['SUMLEV']==50]
df.head()
```

```python
#splitting the data using groupby()

for state, city in df.groupby('STNAME'):
    avg = np.average(city['CENSUS2010POP'])
    print('Cities in ' + state + ' have an average population of ' + str(avg))
```
