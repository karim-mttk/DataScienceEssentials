## Basic Statistical Testing

### Part 1: 

```python
import numpy as np
import pandas as pd
from scipy import stats
from scipy.stats import ttest_ind
```

```python
df = pd.read_csv('datasets/grades.csv')
df.head()
```

```python
#splitting data into two parts: early finishers finished before 2016, late finishers finished after 2016
early_finishers = df[pd.to_datetime(df['assignment1_submission']) < '2016']
early_finishers.head()
```

```python
late_finishers = df[pd.to_datetime(df['assignment1_submission']) > '2016']
late_finishers.head()
```

```python
print('Early finishers avg:', early_finishers['assignment1_grade'].mean())
print('Late finishers avg:', late_finishers['assignment2_grade'].mean())
```

```python
#using a threshold of 5% for alpha.
from scipy.stats import ttest_ind
ttest_ind(early_finishers['assignment1_grade'], late_finishers['assignment1_grade'])
```

From the above the probalitity for assignment is 19% and is higher than the alpha of 5%.
The null hypothesis of two populations being the same cannot be proven by this. So the next assignments can be checked. 

```python
print('assignment2:', ttest_ind(early_finishers['assignment2_grade'], late_finishers['assignment2_grade']))
print('assignment3:', ttest_ind(early_finishers['assignment3_grade'], late_finishers['assignment3_grade']))
print('assignment4:', ttest_ind(early_finishers['assignment4_grade'], late_finishers['assignment4_grade']))
print('assignment5:', ttest_ind(early_finishers['assignment5_grade'], late_finishers['assignment5_grade']))
print('assignment6:', ttest_ind(early_finishers['assignment6_grade'], late_finishers['assignment6_grade']))
```

### Part Two:

```python
#create a dataframe with 100 rows and 100 columns with random numbers
df1 = pd.DataFrame([np.random.random(100) for x in range(100)])
df1.head()
```

```python
df2 = pd.DataFrame([np.random.random(100) for x in range(100)])
df2.head()
```

```python
def test_columns(alpha = 0.1):
    num_diff = 0
    
    for col in df1.columns:
        teststat, pval = ttest_ind(df1[col], df2[col])
        
        if pval<=alpha:
            print('Col {} is statistically different at alpha={} & pval={}'.format(col,alpha,pval))
            num_diff += 1
    print('Number of times it was different: {}'.format(num_diff))        
```

```python
test_columns()
```

```python
#testing with alpha = 0.05
test_columns(0.05)
```
